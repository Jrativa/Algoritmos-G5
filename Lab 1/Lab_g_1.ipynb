```python
def encontrar(s, target): #s, la secuencia; target,la cadena a buscar

  pos = 0                                 #O(1)
  targs = []                              #O(1)
  cad = ""                                #O(1)

#Durante la longitud de la secuencia s
  for i in range(len(s)):                 #O(n)
    if s[i] == target[0]:                 #O(1)

        #Si la actual letra de la cadena es la primera letra del target, vamos a recorrer el target
      for j in range(len(target)):        #O(n)

       #Seguimos recorriendo la secuencia, verificando la igualdad de las letras del target
        if s[i+j]==target[j]:             #O(1)

            #Se van agregando esta letras a cadena
          cad += s[i+j]                   #O(1)

        #Si existe una letra que no coincida con el target se rompe el for y sigue recorriendo la secuencia
        else:                             #O(1)
          cad = ""                        #O(1)
          i+=j                            #O(1)
          break

    #Si la cadena obtenida es igual a target, se agrega a targ, donde se encontro la cadena y hasta donde va
    if cad == target:                           #O(1)
        targs.append((pos,pos+len(target)-1))   #O(1) 
        cad=""                                  #O(1)

        #Acortamos el recorrido del for adelantando la parte donde se encuentra la cadena
        i+=j                                    #O(1)
    pos +=1    
                                     #O(1)
    #Retornamos la lista donde se ve, el comienzo y final donde aparece las cadenas.
  return targs                                  #O(1)

  ''
```
